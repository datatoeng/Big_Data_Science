```
%python
# Rename the columns of df_feature
oldColumns = df_feature.schema.names
newColumns = ['txId', 'time step'] + [f'trans_feat_{i}' for i in range(93)] + [f'agg_feat_{i}' for i in range(72)]

from functools import reduce
 
df_feature_cleaned = reduce(lambda df_feature, idx: df_feature.withColumnRenamed(oldColumns[idx], newColumns[idx]), range(len(oldColumns)), df_feature)
df_feature_cleaned.printSchema()
```

```
# Data join
from pyspark.sql.functions import udf, col
df_data = df_feature_cleaned.join(df_classes, on=['txId'], how='left')

encoding= udf(lambda x: '0' if x == "unknown" else x)
df_data = df_data.withColumn("class",encoding(col("class")))
display(df_data)
```

```
# Check for distinct class
df_data.select("class").distinct().show()
```

```
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.feature import HashingTF, Tokenizer
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
```

```
# "data" only includes the labelled data, which is used to train our ML models
data = df_data.filter((col("class") == "1") | (col("class") == "2")).drop("txId","time step")
# "data_unkown" are the unlabelled records, which will be classified once the ML model is established
data_unknown = df_data.filter(col("class") == "0").drop("time step")
```

```
#Convert the string to float type
from pyspark.sql.functions import col
data_float = data.select([col(column).cast('float') for column in data.columns])
data_unkwnown_float = data_unknown.select([col(column).cast('float') for column in data_unknown.columns])

# Reference: https://towardsdatascience.com/first-time-machine-learning-model-with-pyspark-3684cf406f54
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler

#Collect the features that will be used for ML training
cols = data.columns[0:165]
cols_unknown = data_unknown.columns[1:166]

# features convert into Vector
assembler = VectorAssembler(inputCols=cols,outputCol="features")
transformed_data = assembler.transform(data_float)
assembler_unknown = VectorAssembler(inputCols=cols_unknown,outputCol="features")
transformed_data_unknown = assembler.transform(data_unkwnown_float)
```

```
# train and test data split
(training_data, test_data) = transformed_data.randomSplit([0.8,0.2], seed =2020)
#print("Training Dataset Count: " + str(training_data.count()))
#print("Test Dataset Count: " + str(test_data.count()))
```

```
# Reference: https://towardsdatascience.com/first-time-machine-learning-model-with-pyspark-3684cf406f54
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(labelCol='class', featuresCol='features', maxDepth=5)
model = rf.fit(training_data)
rf_predictions = model.transform(test_data)
```

```
# Evaluate the random forest classifer model
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'class', metricName = 'accuracy')
print('Random Forest classifier Accuracy:', multi_evaluator.evaluate(rf_predictions))
```

```
# Reference: https://towardsdatascience.com/first-time-machine-learning-model-with-pyspark-3684cf406f54
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(featuresCol = 'features', labelCol = 'class', maxIter=10)
lrModel = lr.fit(training_data)
lr_predictions = lrModel.transform(test_data)
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
multi_evaluator = MulticlassClassificationEvaluator(labelCol = 'class', metricName = 'accuracy')
print('Logistic Regression Accuracy:', multi_evaluator.evaluate(lr_predictions))
```
